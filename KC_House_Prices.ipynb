## Building Year and Renovation

The modernisation year and therefore an implied standard can be deducted from building year and the year it was renovated.

```python
# years since last modernisation
kc['modernised'] = datetime.datetime.now().year - kc[[
    'yr_built', 'yr_renovated'
]].max(axis=1)
kc = kc.drop(['yr_built', 'yr_renovated'], axis=1)
```

```python
kc['sqft_basement'] = kc['sqft_living'] - kc['sqft_above']
```

# Features
## Dummies
In order to be able to process the categoria data dummies are created.

```python
dummies = pd.DataFrame()
i_dummies = pd.DataFrame()

for i in cat:
    i_dummies = pd.get_dummies(cat[i], prefix=i, drop_first=True)
    dummies = pd.concat([dummies, i_dummies], axis=1)

kc_dum = pd.concat([cont, dummies], axis=1)
```

Where there is an excessive amount of variables it will be difficult to view them separately. Therefore arrays are defined.

```python
cat.nunique()
```

```python
zipcodes = kc_dum.filter(regex='^zipcode', axis=1)
grades = kc_dum.filter(regex='^grade', axis=1)
```

## Explanatory Variables

```python
X = list(kc_dum)
X.remove('price')
X = [x for x in X if not x.startswith('zipcode')]
X = [x for x in X if not x.startswith('grade')]

rs = []
adj_rs = []

for i in X:
    X1 = kc_dum[i]
    X1 = sm.add_constant(X1)
    y1 = kc_dum.price
    rs.append(sm.OLS(y1, X1).fit().rsquared.round(2))
    adj_rs.append(sm.OLS(y1, X1).fit().rsquared_adj.round(2))

X1 = zipcodes
X1 = sm.add_constant(X1)
y1 = kc_dum.price
rs.append(sm.OLS(y1, X1).fit().rsquared.round(2))
adj_rs.append(sm.OLS(y1, X1).fit().rsquared_adj.round(2))
X.append('zipcodes')

X1 = grades
X1 = sm.add_constant(X1)
y1 = kc_dum.price
rs.append(sm.OLS(y1, X1).fit().rsquared.round(2))
adj_rs.append(sm.OLS(y1, X1).fit().rsquared_adj.round(2))
X.append('grades')

rs_df = pd.DataFrame()
rs_df['explanatory_variable'] = X
rs_df['r_squared'] = rs
rs_df['adjusted_r_squared'] = adj_rs
rs_df = rs_df.sort_values('r_squared', ascending=False)
rs_df
```

# Model Creation

## Using all explanatory variables

```python
X1 = kc_dum.copy()
X1.drop('price', axis=1, inplace=True)
X1 = sm.add_constant(X1)
y1 = kc_dum.price
sm.OLS(y1, X1).fit().summary()
```

The model show good r-squared values, however some variables show insignificance (high p-values).

## Removing insignificant variables

```python
y2 = kc_dum.price
selected_features = list(kc_dum)
selected_features.remove('price')
selected_features.remove('log_price')
pmax = 1
removed_features = []

while (len(selected_features) > 0):
    p = []
    X2 = kc_dum[selected_features]
    X2 = sm.add_constant(X2)
    model = sm.OLS(y2, X2).fit()
    p = pd.Series(model.pvalues.values[1:], index=selected_features)
    pmax = max(p)
    p_max_feature = p.idxmax()
    if (pmax > 0.05):
        selected_features.remove(p_max_feature)
        removed_features.append(p_max_feature)
    else:
        break

print(removed_features)
```

As some of the dummies were removed due to insignificance, the entire group has to be taken out. 

```python
selected_features = [x for x in selected_features if not x.startswith('zipcode')]
selected_features = [x for x in selected_features if not x.startswith('grade')]

X3 = kc_dum[selected_features]
X3 = sm.add_constant(X3)
y3 = kc_dum.price
sm.OLS(y3, X3).fit().summary()
```

## Log Price

```python
y2 = kc_dum.log_price
selected_features = list(kc_dum)
selected_features.remove('price')
selected_features.remove('log_price')
pmax = 1
removed_features = []

while (len(selected_features) > 0):
    p = []
    X2 = kc_dum[selected_features]
    X2 = sm.add_constant(X2)
    model = sm.OLS(y2, X2).fit()
    p = pd.Series(model.pvalues.values[1:], index=selected_features)
    pmax = max(p)
    p_max_feature = p.idxmax()
    if (pmax > 0.05):
        selected_features.remove(p_max_feature)
        removed_features.append(p_max_feature)
    else:
        break

print(removed_features)
```

```python
selected_features = [x for x in selected_features if not x.startswith('zipcode')]
selected_features = [x for x in selected_features if not x.startswith('grade')]

X3 = kc_dum[selected_features]
X3 = sm.add_constant(X3)
y3 = kc_dum.log_price
model = sm.OLS(y3, X3).fit()
model.summary()
```

# Results

```python
X3['valued_at'] = model.predict(X3).round(2)
X3['valued_at'] = [math.exp(float(p)) for p in X3['valued_at']]

result = data.join(X3['valued_at'])
result['valuation_deviation_abs'] = (result['valued_at'] -
                                     result['price']).round(2)
result['valuation_deviation_pct'] = (result['valuation_deviation_abs'] /
                                     result['price'] * 100).round(2)
```

## Valuation

```python
underpriced = result.query('valuation_deviation_pct < -5').shape[0]
overpriced = result.query('valuation_deviation_pct > 5').shape[0]
fairly_priced = result.shape[0] - overpriced - underpriced

valuation = pd.DataFrame({
    'Valuation': ['underpriced', 'overpriced', 'fairly_priced'],
    'No_Houses': [underpriced, overpriced, fairly_priced],
    'Percentage': [
        underpriced / result.shape[0] * 100,
        overpriced / result.shape[0] * 100,
        fairly_priced / result.shape[0] * 100
    ]
})
valuation
```

## Investment Opportunities

```python
plt.scatter(result['price'], result['valued_at']);
plt.show()
```

```python
recs = result[[
    'id', 'price', 'valued_at', 'valuation_deviation_abs',
    'valuation_deviation_pct'
]]
recs = recs.sort_values(by='valuation_deviation_pct',
                        ascending=False)[1:11].round()

recs
```

The objects above are largely underpriced and therefore offer be great opportunities to invest and resell at a substantial profit.
Nonetheless, evaluating the regression output there is potential to improve the prediction model e.g. by exploring further non-linear models and additional features.
